# Grabar y leer Datos en Archivos

```python
import numpy as np
import matplotlib.pyplot as plt
import csv # comma separeted values  7,5
#                                    6,10
#                                    2.3,4.6,4.7,8,1,2.3
```

```python
# Generar datos aleatorios para x e y
np.random.seed(0)
x = np.random.rand(50)
y = 2 * x + 1 + 0.1 * np.random.randn(50)  # y es una funci√≥n lineal de x con ruido
```

```python
print(x, y)
```

```python
# Graficar los datos
plt.scatter(x, y, label='Datos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fico de Datos')
plt.legend()
plt.grid()
plt.show()
```

```python
# Guardar los datos en un archivo CSV
with open('archivito.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['x', 'y'])  # Escribir encabezados
    for i in range(len(x)):
        writer.writerow([x[i], y[i]])
print('terminamos')
```

```python
# Leer los datos desde el archivo CSV
x_leidos = []
y_leidos = []

with open('archivito.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    print("Encabezamiento:", reader.fieldnames)  # Mostrar el encabezado
    for row in reader:
        x_leidos.append(float(row['x']))  # Usar el nombre de la columna en lugar del √≠ndice
        y_leidos.append(float(row['y']))

# Imprimir los datos le√≠dos
for i in range(len(x_leidos)):
    print(f'x: {x_leidos[i]}, y: {y_leidos[i]}')
```

```python
# Graficar los datos nuevamente desde el archivo
plt.figure()
plt.scatter(x_leidos, y_leidos, color='red', label='Datos Le√≠dos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gr√°fico de Datos Le√≠dos')
plt.legend()
plt.grid()
# Mostrar los gr√°ficos
plt.show()
```
