# Markdown

## Markdown

**Markdown** es un lenguaje de *marcado ligero* creado por _John Gruber_ que trata de conseguir la `máxima legibilidad y facilidad de publicación` tanto en su forma de entrada como de salida, inspirándose en muchas convenciones existentes para ***marcar mensajes*** de correo electrónico usando __texto plano__.

## Citando Code
Puede llamar código o un comando dentro de una oración con solo comillas invertidas ` `. El texto dentro de las comillas invertidas no será formateado.

Use `comillas simples invertidas` para enumerar todos los archivos nuevos o modificados que aún no se han confirmado.

## LaTex
Podemos utilizar __LaTex__ para escribir todos los simbolos matemáticos que necesitemos.  
Para ello, solo necesitamos escribir los comandos de __LaTex__ entre dos símbolos __$__ 

Aqui definimos un ejemplo matemático
$\int {\frac{2 x 4}{8.7 - \pi}}$
que quedan muy lindos y dentro del texto.

O tambien podemos mejorarlos, haciéndolos más grandes y centrados, para ello usamos dos símbolos __$__  en cada extremos de la ecuación  
$$\int {\frac{2 x 4}{8.7 - \pi}}$$

O utilizando los comandos __LaTex__  _\\begin{equation}  ...  \\end{equation}_ 

\begin{equation}
\int {\frac{2 x 4}{8.7 - \pi}}
\end{equation}
